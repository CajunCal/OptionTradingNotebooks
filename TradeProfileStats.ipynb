```python
def Chart_Multi_Bins(sample): 
   print('show some stats for features in the set.')
   print(sample.shape)
   fig,axes =plt.subplots(10,3, figsize=(15, 12)) # 3 columns each containing 9 figures, total 27 features
   samplenp = sample.to_numpy()

   ax=axes.ravel()# flat axes with numpy ravel

   print('setting up bins.') 
   for i in range(len(sample.columns.values)):
     _,bins=np.histogram(samplenp[:,i],bins=50)
     
     ax[i].hist(samplenp[:,i],bins=bins,color='r',alpha=.5)# red color 
     ax[i].set_title(sample.columns.values[i],fontsize=11)
     ax[i].axes.get_xaxis().set_visible(False) # the x-axis co-ordinates are not useful. Do not display. 
     ax[i].set_yticks(())
   ax[0].legend(['sample'],loc='best',fontsize=8)
   plt.tight_layout()# let's make good plots
   plt.show()
```
